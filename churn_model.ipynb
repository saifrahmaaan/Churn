```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/Customer-Churn.csv')
```

```python
df.columns = df.columns.str.strip()

print("Column names:", df.columns)
```

```python
X = df.drop(columns=['customerID','Churn'])
Y = df['Churn']
```

```python
X.info()
```

```python
Y.info()
```

```python
X.head()
```

```python
columns_to_drop = ['gender', 'Partner', 'Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod']
```

```python
X_encoded = pd.get_dummies(X, columns=columns_to_drop, drop_first=True)
```

```python
X_encoded.head()
```

```python
columns_to_normalize = ['tenure','MonthlyCharges','TotalCharges']
```

```python
scaler = StandardScaler()
```

```python
# Assuming df_encoded is your DataFrame

# Replace whitespace with NaN and then fill NaN with a default value
X_encoded['TotalCharges'] = pd.to_numeric(X_encoded['TotalCharges'].replace(' ', np.nan), errors='coerce').fillna(0)

# Convert the column to float
X_encoded['TotalCharges'] = X_encoded['TotalCharges'].astype(float)
```

```python
X_encoded[columns_to_normalize] = scaler.fit_transform(X_encoded[columns_to_normalize])
```

```python
X_encoded.info()
```

```python
X_train, X_temp = train_test_split(X_encoded, test_size=0.4, random_state=42)
X_dev, X_test = train_test_split(X_temp, test_size=0.5,random_state=42)
```

```python
X_train.info()
```

```python
X_dev.info()
```

```python
X_test.info()
```

```python
Y.info()
```

```python
label_encoder = LabelEncoder()
Y_encoded = label_encoder.fit_transform(Y)
```

```python
Y_train, Y_temp = train_test_split(Y_encoded, test_size=0.4, random_state=42)
Y_dev, Y_test = train_test_split(Y_temp, test_size=0.5,random_state=42)
```

```python
print(Y_train[:50])
```

```python
# Define the model
model = tf.keras.Sequential([
    layers.Dense(25, activation='relu', input_shape=(30,)),
    layers.Dense(15, activation='relu'),
    layers.Dense(5, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=10, batch_size=32, validation_data=(X_dev, Y_dev))
```

```python
# Assuming your model is already trained and stored in the 'model' variable

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, Y_test)

# Print the test accuracy
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import precision_score
```

```python
Y_pred = model.predict(X_test)
Y_pred_binary = (Y_pred > 0.5).astype(int)
```

```python
precision = precision_score(Y_test, Y_pred_binary)
print(f'Precision: {precision:.2f}')
```

```python
from sklearn.metrics import recall_score

# Assuming y_true and y_pred are your true and predicted labels
recall = recall_score(Y_test, Y_pred_binary)

print("Recall:", recall)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(Y_test, Y_pred_binary)

print("F1 Score:",f1)
```
